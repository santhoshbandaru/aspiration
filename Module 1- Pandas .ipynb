# Query 1.4

In a separate array , calculate the monthwise VWAP (Volume Weighted Average Price ) of the stock. 
( VWAP = sum(price*volume)/sum(volume) ) 
To know more about VWAP , visit - VWAP definition 
{Hint : Create a new dataframe column ‘Month’. The values for this column can be derived from the ‘Date” column by using appropriate pandas functions. Similarly, create a column ‘Year’ and initialize it. Then use the 'groupby()' function by month and year. Finally, calculate the vwap value for each month (i.e. for each group created).

```python
data['Month']=data['Date'].map(lambda x: x.month)
data
```

```python
data['year']=data['Date'].map(lambda x:x.year)
data
```

```python
mydf=data.groupby(['Month','year'])
mydf.describe()
```

```python
data
```

```python
VWAP=[]
def vwap(group, price, volume):
    price= group[price]
    volume= group[volume]
    try:
        VWAP.append((price* volume).sum() / volume.sum())
    except ZeroDivisionError:
        VWAP.append(price.mean())
mydf.apply(vwap,"Close Price","No. of Trades")
VWAP
```

# Query 1.5

Write a function to calculate the average price over the last N days of the stock price data where N is a user defined parameter. Write a second function to calculate the profit/loss percentage over the last N days.
Calculate the average price AND the profit/loss percentages over the course of last -
1 week, 2 weeks, 1 month, 3 months, 6 months and 1 year.
{Note : Profit/Loss percentage between N days is the percentage change between the closing prices of the 2 days }

```python
data1=data.set_index('Date')
```

```python
data1
```

```python
def avg(N):
    mavg= data1['Close Price'].rolling(window=N, center=True,min_periods=1).mean()
    return(mavg)
def pl_pct(N):
    try:
        olddate=data['Date'].max()
        original=data1.loc[olddate,'Close Price']
        difference=pd.Timedelta(days=N)
        newdate=olddate-difference
        new=data1.loc[newdate,'Close Price']
        return abs(((original-new)/original)*100)
    except:
        old=data['Date'].idxmax()
        original=data.loc[old,'Close Price']
        new=data.loc[old-N,'Close Price']
        return abs(((original-new)/original)*100)
    
```

```python
profitloss=[]
movingavg=[]
for i in [7,14,30,90,180,365]:
    profitloss.append(pl_pct(i))
    movingavg.append(avg(i))
print (profitloss)
print(movingavg)
```

# Query 1.7
 Add another column 'Trend' whose values are: 
 'Slight or No change' for 'Day_Perc_Change' in between -0.5 and 0.5 'Slight positive' for 'Day_Perc_Change' in between 0.5 and 1 
 'Slight negative' for 'Day_Perc_Change' in between -0.5 and -1 'Positive' for 'Day_Perc_Change' in between 1 and 3 
 'Negative' for 'Day_Perc_Change' in between -1 and -3
 'Among top gainers' for 'Day_Perc_Change' in between 3 and 7 
 'Among top losers' for 'Day_Perc_Change' in between -3 and -7 
 'Bull run' for 'Day_Perc_Change' >7 
 'Bear drop' for 'Day_Perc_Change' <-7

```python
data1.reset_index(inplace=True)
```

```python
data1['Trend']='default value'
for index in range(len(data1)):
    if -0.5<data1.loc[index,"Day_Perc_Change"]<0.5:
        data1.loc[index,"Trend"]='Slight or No change'
    elif 0.5<data1.loc[index,"Day_Perc_Change"]<1:
        data1.loc[index,"Trend"]='Slight Positive'
    elif -1<data1.loc[index,"Day_Perc_Change"]<-0.5:
        data1.loc[index,"Trend"]='Slight Negative'
    elif 1<data1.loc[index,"Day_Perc_Change"]<3:
        data1.loc[index,"Trend"]='Positive'
    elif -3<data1.loc[index,"Day_Perc_Change"]<-1:
        data1.loc[index,"Trend"]='Negative'
    elif 3<data1.loc[index,"Day_Perc_Change"]<7:
        data1.loc[index,"Trend"]='Among top gainers'
    elif -7<data1.loc[index,"Day_Perc_Change"]<-3:
        data1.loc[index,"Trend"]='Among top losers'
    elif data1.loc[index,"Day_Perc_Change"]>7:
        data1.loc[index,"Trend"]='Bull run'
    elif data1.loc[index,"Day_Perc_Change"]<-7:
        data1.loc[index,"Trend"]='Bear drop'
data1
```
